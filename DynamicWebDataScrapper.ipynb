```python
import requests
import pandas as pd
import time
```

```python
def initialize_excel():
    url = "https://oilprice.com/oil-price-charts/#prices "

    r = requests.get(url)
    df_list = pd.read_html(r.text) # this parses all the tables in webpages to a list
    data = df_list[0]

    table = data.iloc[:25 , [1, 2, 3, 4]]
#     print(table)
    
#     After parsing the data we see the data fetched are not properly ready to be fed into the excel file the way we like
#     to be hence we shall use the following lines of code to process the data to our need

    table.columns = ['Name', 'Price', 'Change', 'Four']
    table[['ChangePercentage', 'Delay']] = table['Four'].str.split('(', expand=True)
#    table['ChangePercentage'] = table['ChangePercentage'].str.replace('[\+\-%]', '', regex=True) 

#    table['Delay'] = table['Delay'].str.replace('[()%]', '') # Removes the brackets from the 'Delay' column
   
    table.drop(columns=['Four'], inplace=True) # Drop the original column 4
    table['Price'] = table['Price'].astype(float)
    table['Change'] = table['Change'].astype(float)
    
    table.to_excel("oil_price.xlsx", index=False)
```

```python
# We shall fetch oil price data to call for updating the table
```

```python
def fetch_oil_price_data():
    url = "https://oilprice.com/oil-price-charts/#prices "

    r = requests.get(url)
    df_list = pd.read_html(r.text)
    data = df_list[0]

    table = data.iloc[:25 , [1, 2, 3, 4]]

    table.columns = ['Name', 'Price', 'Change', 'Four']
    table[['ChangePercentage', 'Delay']] = table['Four'].str.split('(', expand=True)
#     table['ChangePercentage'] = table['ChangePercentage'].str.replace('[\+\-%]', '', regex=True)

    
#     table['Delay'] = table['Delay'].str.replace('[()%]', '') # Remove the brackets from the 'Delay' column

    # Drop the original column 4
    table.drop(columns=['Four'], inplace=True)
    table['Price'] = table['Price'].astype(float)
    table['Change'] = table['Change'].astype(float)

    return table
```

```python
def update_excel(oil_data, file_path):
    try:
        # Loads existing data from the Excel file
        df = pd.read_excel(file_path)

        # Converts 'Price' and 'Change' columns to float in the existing DataFrame
        df['Price'] = df['Price'].astype(float)
        df['Change'] = df['Change'].astype(float)

        # Checks if the fetched data is different from the existing data
        if df.equals(oil_data):
            print("No changes in the data. Not updating the Excel file.")
        else:
            
            oil_data.to_excel(file_path, index=False)

            print("Data updated successfully.")
    except Exception as e:
        print("Failed to update Excel file:", e)
```

```python
if __name__ == "__main__":
    file_path = "E:\Intern RPA\oil_price.xlsx"  # Replace with the desired file path

    update_interval = 20  # Update interval in 20 seconds
    total_duration = 300  # Total duration for checking updates (5 minutes)
    elapsed_time = 0

    while elapsed_time <= total_duration:
        start_time = time.time()

        oil_data = fetch_oil_price_data()
        if oil_data is not None and not oil_data.empty:
            update_excel(oil_data, file_path)
            print(oil_data)

        elapsed_time += time.time() - start_time  # Add the total time taken for the loop iteration
        time.sleep(update_interval)
        print(elapsed_time)

    print("Data update stopped after 5 minutes.")
```
